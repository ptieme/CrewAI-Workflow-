## Agents

```python
from crewai import Agent
from crewai_tools import WebsiteSearchTool, SerperDevTool, FileReadTool

web_search_tool = WebsiteSearchTool()
serper_dev_tool = SerperDevTool()
file_read_tool = FileReadTool(
    file_path = 'exemple_description_job.md',
    description = 'Outil pour lire le fichier exemple de description de poste'
)
```

```python
class Agents:
    # AGENT 1 : L'ANALYSTE
    def research_agent(self):
        return Agent(
            role = "Analyste",
            goal = """Analyser le site web de l'entreprise pour comprendre sa culture,
                    ses valeurs et ses besoins spécifiques.""",
            tools = [web_search_tool, serper_dev_tool],  # Pas besoin de lire le template ici
            max_iter = 6,
            backstory = """C'est un expert pour décrypter la culture d'une entreprise à partir
                        de son site web et de courtes descriptions""",
            verbose = True
        )

    # AGENT 2 : LE RÉDACTEUR
    def writer_agent(self):
        return Agent(
            role = "Rédacteur d'offres d'emploi",
            goal = """Utiliser les informations de l'Analyste pour écrire une offre d'emploi détaillée,
                    intéressante et qui donne envie de postuler.""",
            tools = [web_search_tool, serper_dev_tool, file_read_tool],
            max_iter = 4,
            backstory = """C'est un pro de la rédaction, capable de créer des offres d'emploi
                    qui parlent aux bons candidats et qui reflètent l'esprit de l'entreprise.""",
            verbose = True
        )

    # AGENT 3 : LE CORRECTEUR
    def review_agent(self):
        return Agent(
            role = "Spécialiste de la Relecture",
            goal = """Vérifier que l'offre d'emploi est claire, attractive, sans fautes de grammaire,
                     et bien alignée avec les valeurs de l'entreprise. Il l'améliore pour qu'elle soit parfaite.""",
            tools = [web_search_tool, serper_dev_tool, file_read_tool],
            max_iter = 3,
            backstory = """C'est un relecteur très minutieux avec un œil de lynx.
                        Il s'assure que tout contenu est impeccable, clair et engageant.""",
            verbose = True
        )
```

## Tasks

```python
from textwrap import dedent
from crewai import Task

class Tasks:

    # ------------------------------------------------------------------ #
    #    Tâche 1 : Recherche sur l'entreprise
    #    → confiée à l’Analyste
    # ------------------------------------------------------------------ #
    def company_culture_task(self, agent, company_url, company_description):
        return Task(
            description = dedent(f"""\
                Analyse le site web ({company_url}) et la description de l'entreprise "{company_description}"
                pour bien comprendre sa culture, ses valeurs et sa mission.

                Trouve ce qui rend cette entreprise unique (ses projets, ses réussites)
                et compile un rapport pour le Rédacteur.
            """),
            expected_output = dedent("""\
                Un rapport qui résume la culture, les valeurs et les points forts de l'entreprise,
                avec des suggestions pour les intégrer dans l'offre d'emploi.
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    Tâche 2 : Recherche sur les besoins du poste
    #    → confiée à l'Analyste
    # ------------------------------------------------------------------ #
    def role_requirements_task(self, agent, hiring_needs):
        return Task(
            description = dedent(f"""\
                En te basant sur les besoins du recruteur : {hiring_needs}, identifie les compétences,
                l'expérience et les qualités idéales pour ce poste.

                Pense aux projets en cours de l'entreprise et aux tendances du secteur.
            """),
            expected_output = dedent("""\
                Une liste des compétences, expériences et qualités recommandées pour le candidat idéal,
                en accord avec la culture de l'entreprise et les besoins du poste.
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    Tâche 3 : Rédaction de la première version
    #    → confiée au Rédacteur
    # ------------------------------------------------------------------ #
    def draft_job_posting_task(self, agent, company_description, hiring_needs, specific_benefits):
        return Task(
            description = dedent(f"""\
                Rédige une première version de l'offre d'emploi pour le poste requis par le manager : {hiring_needs}.
                Commence par une introduction accrocheuse (en prenant en compte la description de l'entreprise - {company_description}),
                puis détaille le poste, les responsabilités et les compétences requises.

                Le ton doit correspondre à la culture de l'entreprise.
                N'oublie pas d'inclure les avantages spécifiques proposés : {specific_benefits}
            """),
            expected_output = dedent("""\
                Une offre d'emploi détaillée et engageante, avec une introduction, une description du poste,
                les responsabilités, les compétences requises et les avantages uniques de l'entreprise.

                (Format Markdown selon le modèle du fichier exemple_description_job.md - # Titre / # Sous-titre)
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    Tâche 4 : Relecture et finalisation de l'offre
    #    → confiée au Correcteur
    # ------------------------------------------------------------------ #
    def review_job_posting_task(self, agent, hiring_needs):
        return Task(
            description = dedent(f"""\
                Relis la version de l'offre d'emploi pour le poste suivant : {hiring_needs}.
                Vérifie qu'elle est claire, attractive, sans fautes, et qu'elle correspond
                bien à l'esprit de l'entreprise.

                Apporte les modifications nécessaires pour qu'elle soit parfaite.
            """),
            expected_output = dedent("""\
                Une offre d'emploi finale, sans erreur, claire, engageante et parfaitement alignée avec la culture de l'entreprise.
                Le tout, formaté en Markdown et prêt à être publié dans le fichier offre_emploi.md.
            """),
            agent = agent,
            output_file = "offre_emploi.md"
        )

    # ------------------------------------------------------------------ #
    #    Tâche 5 : Analyse du secteur d'activité
    #    → confiée à l'Analyste
    # ------------------------------------------------------------------ #
    def industry_analysis_task(self, agent, company_url, company_description):
        return Task(
            description = dedent(f"""\
                Fais une analyse du secteur d'activité de l'entreprise
                - Site web de l'entreprise : {company_url})
                - Description de l'entreprise : {company_description}

                Étudie les tendances actuelles, les défis et les opportunités.
                Explique comment le poste à pourvoir peut aider l'entreprise à s'adapter à ces tendances et à se démarquer.
            """),
            expected_output = dedent("""\
                Un rapport d'analyse sur les tendances du secteur. Ce rapport doit donner des idées
                pour présenter le poste et l'entreprise comme un choix très attractif pour les meilleurs candidats.
            """),
            agent = agent
        )
```

## Crew

```python
print("Bienvenue dans l'assistant de création d'offres d'emploi !")
print("Veuillez répondre aux questions suivantes pour générer une annonce parfaite.")

# --- Étape 1: Infos sur l'entreprise ---
print("\n--- 1. À propos de l'entreprise ---")
company_url = input(
    "Quel est le site web de l'entreprise ? (ex: www.stripe.com)\n"
    "-> L'IA analysera le site pour comprendre la culture et la mission.\n> "
)

company_description = input(
    "En une phrase, quel est le cœur de métier de l'entreprise ? (ex: 'Plateforme de paiement en ligne pour les entreprises')\n> "
)

# --- Étape 2: Le Poste ---
print("\n--- 2. Description du poste ---")
hiring_needs = input(
    "Décrivez précisément le profil recherché.\n"
    "Ex: 'Développeur Full-Stack (React/Node.js) confirmé pour l'équipe 'Growth', avec 3+ ans d'expérience. "
    "Il/Elle travaillera sur l'optimisation du parcours utilisateur.'\n> "
)

# --- Étape 3: Les Avantages ---
print("\n--- 3. Avantages et arguments de vente ---")
specific_benefits = input(
    "Listez les avantages uniques et concrets que vous offrez.\n"
    "Ex: 'Télétravail flexible (2 jours/semaine), RTT, budget matériel de 1500€, tickets restaurant (Swile), bureaux au centre-ville.'\n> "
)

print("\nMerci !")
```

```python
# Création des agents
agents = Agents()

researcher_agent = agents.research_agent()
writer_agent = agents.writer_agent()
review_agent = agents.review_agent()
```

```python
# Création des tâches
tasks = Tasks()


company_culture_task = tasks.company_culture_task(
    researcher_agent,
    company_url,
    company_description
)

industry_analysis_task = tasks.industry_analysis_task(
    researcher_agent,
    company_description,
    company_url
)

role_requirements_task = tasks.role_requirements_task(
    researcher_agent,
    hiring_needs
)

draft_job_posting_task = tasks.draft_job_posting_task(
    writer_agent,
    company_description,
    hiring_needs,
    specific_benefits
)

review_job_posting_task = tasks.review_job_posting_task(
    review_agent,
    hiring_needs
)
```

```python
from crewai import Crew

# Créer un Crew à l'aide d'un processus séquentiel
crew = Crew(
    agents = [researcher_agent, writer_agent, review_agent],
    tasks = [
        company_culture_task,
        industry_analysis_task,
        role_requirements_task,
        draft_job_posting_task,
        review_job_posting_task
    ]
)
```

```python
result = crew.kickoff()
print(result)
```
