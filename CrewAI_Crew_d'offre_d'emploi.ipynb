## Agents

```python
from crewai import Agent
from crewai_tools import WebsiteSearchTool, SerperDevTool, FileReadTool

web_search_tool = WebsiteSearchTool()
serper_dev_tool = SerperDevTool()
file_read_tool = FileReadTool(
    file_path = 'exemple_description_job.md',
    description = 'Outil pour lire le fichier exemple de description de poste'
)
```

```python
class Agents:
    # AGENT 1 : L'ANALYSTE
    def research_agent(self):
        return Agent(
            role = "Analyste",
            goal = """Analyser le site web de l'entreprise pour comprendre sa culture,
                    ses valeurs et ses besoins spÃ©cifiques.""",
            tools = [web_search_tool, serper_dev_tool],  # Pas besoin de lire le template ici
            max_iter = 6,
            backstory = """C'est un expert pour dÃ©crypter la culture d'une entreprise Ã  partir
                        de son site web et de courtes descriptions""",
            verbose = True
        )

    # AGENT 2 : LE RÃ‰DACTEUR
    def writer_agent(self):
        return Agent(
            role = "RÃ©dacteur d'offres d'emploi",
            goal = """Utiliser les informations de l'Analyste pour Ã©crire une offre d'emploi dÃ©taillÃ©e,
                    intÃ©ressante et qui donne envie de postuler.""",
            tools = [web_search_tool, serper_dev_tool, file_read_tool],
            max_iter = 4,
            backstory = """C'est un pro de la rÃ©daction, capable de crÃ©er des offres d'emploi
                    qui parlent aux bons candidats et qui reflÃ¨tent l'esprit de l'entreprise.""",
            verbose = True
        )

    # AGENT 3 : LE CORRECTEUR
    def review_agent(self):
        return Agent(
            role = "SpÃ©cialiste de la Relecture",
            goal = """VÃ©rifier que l'offre d'emploi est claire, attractive, sans fautes de grammaire,
                     et bien alignÃ©e avec les valeurs de l'entreprise. Il l'amÃ©liore pour qu'elle soit parfaite.""",
            tools = [web_search_tool, serper_dev_tool, file_read_tool],
            max_iter = 3,
            backstory = """C'est un relecteur trÃ¨s minutieux avec un Å“il de lynx.
                        Il s'assure que tout contenu est impeccable, clair et engageant.""",
            verbose = True
        )
```

## Tasks

```python
from textwrap import dedent
from crewai import Task

class Tasks:

    # ------------------------------------------------------------------ #
    #    TÃ¢che 1 : Recherche sur l'entreprise
    #    â†’ confiÃ©e Ã  lâ€™Analyste
    # ------------------------------------------------------------------ #
    def company_culture_task(self, agent, company_url, company_description):
        return Task(
            description = dedent(f"""\
                Analyse le site web ({company_url}) et la description de l'entreprise "{company_description}"
                pour bien comprendre sa culture, ses valeurs et sa mission.

                Trouve ce qui rend cette entreprise unique (ses projets, ses rÃ©ussites)
                et compile un rapport pour le RÃ©dacteur.
            """),
            expected_output = dedent("""\
                Un rapport qui rÃ©sume la culture, les valeurs et les points forts de l'entreprise,
                avec des suggestions pour les intÃ©grer dans l'offre d'emploi.
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    TÃ¢che 2 : Recherche sur les besoins du poste
    #    â†’ confiÃ©e Ã  l'Analyste
    # ------------------------------------------------------------------ #
    def role_requirements_task(self, agent, hiring_needs):
        return Task(
            description = dedent(f"""\
                En te basant sur les besoins du recruteur : {hiring_needs}, identifie les compÃ©tences,
                l'expÃ©rience et les qualitÃ©s idÃ©ales pour ce poste.

                Pense aux projets en cours de l'entreprise et aux tendances du secteur.
            """),
            expected_output = dedent("""\
                Une liste des compÃ©tences, expÃ©riences et qualitÃ©s recommandÃ©es pour le candidat idÃ©al,
                en accord avec la culture de l'entreprise et les besoins du poste.
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    TÃ¢che 3 : RÃ©daction de la premiÃ¨re version
    #    â†’ confiÃ©e au RÃ©dacteur
    # ------------------------------------------------------------------ #
    def draft_job_posting_task(self, agent, company_description, hiring_needs, specific_benefits):
        return Task(
            description = dedent(f"""\
                RÃ©dige une premiÃ¨re version de l'offre d'emploi pour le poste requis par le manager : {hiring_needs}.
                Commence par une introduction accrocheuse (en prenant en compte la description de l'entreprise - {company_description}),
                puis dÃ©taille le poste, les responsabilitÃ©s et les compÃ©tences requises.

                Le ton doit correspondre Ã  la culture de l'entreprise.
                N'oublie pas d'inclure les avantages spÃ©cifiques proposÃ©s : {specific_benefits}
            """),
            expected_output = dedent("""\
                Une offre d'emploi dÃ©taillÃ©e et engageante, avec une introduction, une description du poste,
                les responsabilitÃ©s, les compÃ©tences requises et les avantages uniques de l'entreprise.

                (Format Markdown selon le modÃ¨le du fichier exemple_description_job.md - # Titre / # Sous-titre)
            """),
            agent = agent
        )

    # ------------------------------------------------------------------ #
    #    TÃ¢che 4 : Relecture et finalisation de l'offre
    #    â†’ confiÃ©e au Correcteur
    # ------------------------------------------------------------------ #
    def review_job_posting_task(self, agent, hiring_needs):
        return Task(
            description = dedent(f"""\
                Relis la version de l'offre d'emploi pour le poste suivant : {hiring_needs}.
                VÃ©rifie qu'elle est claire, attractive, sans fautes, et qu'elle correspond
                bien Ã  l'esprit de l'entreprise.

                Apporte les modifications nÃ©cessaires pour qu'elle soit parfaite.
            """),
            expected_output = dedent("""\
                Une offre d'emploi finale, sans erreur, claire, engageante et parfaitement alignÃ©e avec la culture de l'entreprise.
                Le tout, formatÃ© en Markdown et prÃªt Ã  Ãªtre publiÃ© dans le fichier offre_emploi.md.
            """),
            agent = agent,
            output_file = "offre_emploi.md"
        )

    # ------------------------------------------------------------------ #
    #    TÃ¢che 5 : Analyse du secteur d'activitÃ©
    #    â†’ confiÃ©e Ã  l'Analyste
    # ------------------------------------------------------------------ #
    def industry_analysis_task(self, agent, company_url, company_description):
        return Task(
            description = dedent(f"""\
                Fais une analyse du secteur d'activitÃ© de l'entreprise
                - Site web de l'entreprise : {company_url})
                - Description de l'entreprise : {company_description}

                Ã‰tudie les tendances actuelles, les dÃ©fis et les opportunitÃ©s.
                Explique comment le poste Ã  pourvoir peut aider l'entreprise Ã  s'adapter Ã  ces tendances et Ã  se dÃ©marquer.
            """),
            expected_output = dedent("""\
                Un rapport d'analyse sur les tendances du secteur. Ce rapport doit donner des idÃ©es
                pour prÃ©senter le poste et l'entreprise comme un choix trÃ¨s attractif pour les meilleurs candidats.
            """),
            agent = agent
        )
```

## Crew

```python
print("Bienvenue dans l'assistant de crÃ©ation d'offres d'emploi !")
print("Veuillez rÃ©pondre aux questions suivantes pour gÃ©nÃ©rer une annonce parfaite.")

# --- Ã‰tape 1: Infos sur l'entreprise ---
print("\n--- 1. Ã€ propos de l'entreprise ---")
company_url = input(
    "Quel est le site web de l'entreprise ? (ex: www.stripe.com)\n"
    "-> L'IA analysera le site pour comprendre la culture et la mission.\n> "
)

company_description = input(
    "En une phrase, quel est le cÅ“ur de mÃ©tier de l'entreprise ? (ex: 'Plateforme de paiement en ligne pour les entreprises')\n> "
)

# --- Ã‰tape 2: Le Poste ---
print("\n--- 2. Description du poste ---")
hiring_needs = input(
    "DÃ©crivez prÃ©cisÃ©ment le profil recherchÃ©.\n"
    "Ex: 'DÃ©veloppeur Full-Stack (React/Node.js) confirmÃ© pour l'Ã©quipe 'Growth', avec 3+ ans d'expÃ©rience. "
    "Il/Elle travaillera sur l'optimisation du parcours utilisateur.'\n> "
)

# --- Ã‰tape 3: Les Avantages ---
print("\n--- 3. Avantages et arguments de vente ---")
specific_benefits = input(
    "Listez les avantages uniques et concrets que vous offrez.\n"
    "Ex: 'TÃ©lÃ©travail flexible (2 jours/semaine), RTT, budget matÃ©riel de 1500â‚¬, tickets restaurant (Swile), bureaux au centre-ville.'\n> "
)

print("\nMerci !")
```

```python
# CrÃ©ation des agents
agents = Agents()

researcher_agent = agents.research_agent()
writer_agent = agents.writer_agent()
review_agent = agents.review_agent()
```

```python
# CrÃ©ation des tÃ¢ches
tasks = Tasks()


company_culture_task = tasks.company_culture_task(
    researcher_agent,
    company_url,
    company_description
)

industry_analysis_task = tasks.industry_analysis_task(
    researcher_agent,
    company_description,
    company_url
)

role_requirements_task = tasks.role_requirements_task(
    researcher_agent,
    hiring_needs
)

draft_job_posting_task = tasks.draft_job_posting_task(
    writer_agent,
    company_description,
    hiring_needs,
    specific_benefits
)

review_job_posting_task = tasks.review_job_posting_task(
    review_agent,
    hiring_needs
)
```

```python
from crewai import Crew

# CrÃ©er un Crew Ã  l'aide d'un processus sÃ©quentiel
crew = Crew(
    agents = [researcher_agent, writer_agent, review_agent],
    tasks = [
        company_culture_task,
        industry_analysis_task,
        role_requirements_task,
        draft_job_posting_task,
        review_job_posting_task
    ]
)
```

```python
result = crew.kickoff()
print(result)
```
